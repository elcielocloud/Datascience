<h2> 23년 1월을 예제로 데이터를 어떻게 가공하는지 보여주는 예시

```python
import pandas as pd
rental_2301 = pd.read_csv('서울특별시 공공자전거 이용정보(시간대별)_2301.csv', encoding='cp949')
sample = pd.read_csv('[상위 30개]2023년 따릉이 대여소별 대여반납건수(반납건수-대여건수 내림차순)_240512.csv', encoding='cp949')
```

```python
# 23년 1월의 공공자전거 이용정보 데이터 확인
rental_2301
```

```python
# 표본으로 활용할 반납-대여가 원활히 되지 않는 상위 30개 대여소 데이터 확인
sample
```

```python
# 23년 1월 데이터에서 필요하지 않은 변수 제거
rental_2301.drop(['대여구분코드', '성별', '연령대코드', '운동량', '탄소량', '이동거리(M)', '이용시간(분)'], axis=1, inplace=True)
rental_2301
```

```python
# 표본 대여소만 활용하기 위한 데이터 처리
# 표본 대여소의 대여소 번호 목록을 담은 리스트 생성
sample_num = sorted(list(sample['대여소번호']))
del_idx_list = []

# 23년 1월 데이터에서 제거할 인덱스 추출
for idx in rental_2301.index:
    rental_num = rental_2301.loc[idx]['대여소번호']
    if rental_num not in sample_num:
        del_idx_list.append(idx)
```

```python
# 표본으로 취할 대여소의 대여소번호 목록
sample_num
```

```python
# 23년 1월 데이터에서 표본 대여소 외 대여소 정보 삭제
rental_2301.drop(del_idx_list, inplace=True)
```

```python
#대여소명 (233. 정류소이름) -> (정류소이름) 으로 변경
new_name = []

for name in list(rental_2301['대여소명']):
    # '.'을 기준으로 분리 후 정류소 이름만 취하기
    name = name.split('.')[1]
    # 앞뒤 공백 제거
    name = name.strip()
    new_name.append(name)
```

```python
# 대여소명 열 이름만 남도록 수정
rental_2301['대여소명'] = new_name
```

```python
# 대여소번호와 대여소명이 분리된 것 확인
rental_2301
```

```python
list(rental_2301['대여소명'])
```

```python
# 대여일자 - 대여시간 - 대여소번호 별로 정렬
rental_2301.sort_values(by=['대여일자', '대여시간', '대여소번호'], inplace=True)
```

```python
# 인덱스 초기화
rental_2301 = rental_2301.reset_index(drop=True)
```

```python
rental_2301
```

```python
# 대여일자 순환 위한 날짜 리스트 생성
month_date = []

for i in range(1, 32):
    month = '2023-01-'
    if i < 10:
        date = '0' + str(i)
    else:
        date = str(i)
    month_date.append(month + date)
```

```python
month_date
```

```python
# 대여소별 시간대별 이용건수 집계를 위한 새로운 데이터프레임 생성
# 변수 : 대여일자 - 대여시간대 - 대여소번호 - 대여소명 - 이용건수

station_rental_2301 = pd.DataFrame(columns=['대여일자', '대여시간', '대여소번호', '대여소명', '이용건수'])
station_rental_2301
```

```python
# 집계할 데이터 프레임의 열, 행 생성
# 대여소별 일-시간대-이용건수 생성 : 이용건수는 모두 0으로 설정 (추후 23년 1월 데이터를 보며 추가)
# 표본 대여소 별로 날짜 - 대여시간 - 이용건수가 담긴 데이터 생성

# idx : 집계 데이터의 인덱스
idx = 0

for date in month_date:
    for s_index in sample.index:
        for time in range(24):
            new_line = [date, time, sample.loc[s_index]['대여소번호'], sample.loc[s_index]['대여소명'], 0]
            station_rental_2301.loc[idx] = new_line
            idx += 1    
```

```python
station_rental_2301
```

```python
# 23년 01월 대여정보를 통해 집계 데이터프레임에 대여건수 추가

# 시간대별 이용건수 데이터프레임을 순환하며 이용건수를 대여소+시간대별 데이터프레임에 추가
for idx in rental_2301.index:
    info = rental_2301.loc[idx]
    date = info['대여일자']
    time = info['대여시간']
    station_num = info['대여소번호']
    # staion_name = info['대여소명'] 은 불필요
    rental_num = info['이용건수']
    
    # 수정 (이용건수 추가할) 인덱스를 위 조건에 따라 추출
    modify_index = station_rental_2301[(station_rental_2301['대여일자'] == date) & (station_rental_2301['대여시간'] == time) & (station_rental_2301['대여소번호'] == station_num)].index
    # 수정 인덱스의 이용건수에 추가
    station_rental_2301.loc[modify_index, '이용건수'] += rental_num
    
```

```python
# 이용건수 정보가 추가됨
station_rental_2301
```

```python
# 평일인지, 평일이 아닌지(주말, 공휴일)는 대여건수에 영향을 미치는 변수
# 대여일자가 평일인지 주말인지 datetime의 weekday() 함수를 통해 요일을 구분
# 0, 1, 2, 3, 4 = 월, 화, 수, 목, 금 = 평일
# 5, 6 = 토, 일 = 주말

import datetime

# 평일구분 열 추가 -> 값 : '평일' or '평일 외'가 될 것

# 빈 열 생성
station_rental_2301['평일구분'] = ''

for idx in station_rental_2301.index:
    date = station_rental_2301.loc[idx]['대여일자']
    # 기존 문자열을 datetime 변수로 변환
    convert_date = datetime.datetime.strptime(date, '%Y-%m-%d')
    # 요일 반환 (0~6)
    day = convert_date.weekday()
    if day < 5:
        # 평일이면 '평일'로 값 변경
        station_rental_2301.loc[idx, '평일구분'] = '평일'
    else:
        # 주말이면 '평일 외'로 값 변경, 추후 공휴일도 '평일 외'로 분류 예정
        station_rental_2301.loc[idx, '평일구분'] = '평일 외'
```
