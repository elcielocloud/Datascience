```python
import pandas as pd

# 23년 월별 대여소별 이용정보 데이터 불러오기
data_2023_1 = pd.read_csv('서울특별시 공공자전거 대여소별 이용정보(월별)_23.1-6.csv', encoding='cp949')
data_2023_2 = pd.read_csv('서울특별시 공공자전거 대여소별 이용정보(월별)_23.7-12.csv', encoding='cp949')
```

```python
# 2023년 하반기 대여소 목록 확인
# 상반기랑 대여소 수가 다름
data_2023_2['대여소명']
```

- 1년의 데이터를 분석해야 하므로 2023년 사이에 사라지거나 새로 생긴 대여소는 포함하지 않기로 함
- 2023년 상반기, 하반기에 모두 존재하는 대여소만 추출하기

```python
cyclelist_2023_1 = list(data_2023_1['대여소명'])
cyclelist_2023_2 = list(data_2023_2['대여소명'])
```

```python
# 대여소명을 대여소 번호와 대여소명으로 분리 (상반기)
rental_num = []
rental_name = []

for i in cyclelist_2023_1:
    splited_info = i.split('.')
    if '에이텍' in i:
        # 에이택(as센터)는 대여소가 아니므로 추후 제외하기 위해, 대여소번호 9999로 임시 설정
        rental_num.append(9999)
        rental_name.append('AS센터')
    else:
        rental_num.append(int(splited_info[0]))
        rental_name.append(splited_info[1].strip())
    
```

```python
data_2023_1['대여소명'] = rental_name
data_2023_1['대여소번호'] = rental_num
data_2023_1
```

```python
# 대여소명을 대여소 번호와 대여소명으로 분리 (하반기)
rental_num2 = []
rental_name2 = []

for i in cyclelist_2023_2:
    splited_info = i.split('.')
    if i == 'AS센터':
        # as센터는 대여소가 아니므로 추후 제외하기 위해, 대여소번호 9999로 임시 설정
        rental_num2.append(9999)
        rental_name2.append('AS센터')
    else:
        rental_num2.append(int(splited_info[0]))
        rental_name2.append(splited_info[1].strip())
    
```

```python
data_2023_2['대여소명'] = rental_name2
data_2023_2['대여소번호'] = rental_num2
data_2023_2
```

```python
# 2023 상/하반기 데이터프레임 합치기
data_2023 = pd.concat([data_2023_1, data_2023_2])
data_2023
```

```python
# 대여소번호, 기준년월 기준으로 데이터프레임 정렬
data_2023 = data_2023.sort_values(by=['대여소번호', '기준년월'])
data_2023 = data_2023.reset_index(drop=True)
data_2023
```

```python
# AS센터 (대여소번호 9999로 설정한) 삭제
data_2023.drop(data_2023[data_2023['대여소번호'] == 9999].index, inplace=True)
```

```python
# 대여소번호 9979 테스트 대여소 추가 삭제
data_2023.drop(data_2023[data_2023['대여소번호'] == 9979].index, inplace=True)
data_2023
```

```python
# 대여건수 중 문자열 정수형으로 바꾸기
for i in data_2023.index:
    number = data_2023['대여건수'].loc[i]
    if isinstance(number, str):
        number = number.strip()
        if ',' in number:
            number = number.replace(',', '')
        if '-' in number:
            number = '0'
        number = int(number)
        data_2023['대여건수'].loc[i] = number
```

```python
# 반납건수 중 문자열 정수형으로 바꾸기
for i in data_2023.index:
    number = data_2023['반납건수'].loc[i]
    if isinstance(number, str):
        number = number.strip()
        if ',' in number:
            number = number.replace(',', '')
        if '-' in number:
            number = '0'
        number = int(number)
        data_2023['반납건수'].loc[i] = number
```

```python
#대여소 번호 목록(집합)
rental_num_2023 = set(data_2023['대여소번호'])
len(rental_num_2023)
```

```python
data_2023
```

```python

#새 데이터프레임 (대여반납 순위)에 넣을 열 자료들
rent_num = []
rent_name = []
rent_area = []
borrow = []
turnin = []
collected = []

for num in list(rental_num_2023):
    #대여소번호, 대여소명, 자치구 추가
    rent_num.append(num)
    
    num_name = list(data_2023[data_2023['대여소번호'] == num]['대여소명'])[-1] # 최신 이름으로
    rent_name.append(num_name)
    
    num_area = list(data_2023[data_2023['대여소번호'] == num]['자치구'])[-1]
    rent_area.append(num_area)
    
    #대여소별 1년 대여건수, 반납건수 합, 대여건수-반납건수
    num_borrow = sum(data_2023[data_2023['대여소번호'] == num]['대여건수'])
    num_turnin = sum(data_2023[data_2023['대여소번호'] == num]['반납건수'])
    num_collect = num_turnin - num_borrow
    
    borrow.append(num_borrow)
    turnin.append(num_turnin)
    collected.append(num_collect)
    
    
    
```

```python
# 대여소별 1년 대여건수/반납건수와 반납-대여건수 데이터가 담긴 새로운 데이터프레임 생성
data_rank_2023 = pd.DataFrame({'대여소번호' : rent_num,
                               '대여소명' : rent_name,
                               '자치구' : rent_area,
                               '대여건수' : borrow,
                               '반납건수' : turnin,
                               '반납건수-대여건수' : collected})

data_rank_2023
```

```python
# 반납-대여건수 내림차순으로 정렬 후 내보내기
data_rank_2023_sorted2 = data_rank_2023.sort_values(by='반납건수-대여건수')
data_rank_2023_sorted2.to_csv('/Users/parkhaneul/DataScience/project_datascience/2023년 따릉이 대여소별 대여반납건수(반납건수-대여건수 내림차순)_240512.csv', index=False, encoding='cp949')
```
